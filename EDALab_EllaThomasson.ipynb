Part 4

```python
# Statistical description
description = airbnb['Price'].describe()
print("Statistical Description:")
print(description)
```

```python
# histogram
sns.histplot(airbnb['Price'], bins=30, kde=False, color='blue')
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
# Kernel Density Estimate (KDE)
sns.kdeplot(airbnb['Price'], fill=True, color='orange')
plt.title('Kernel Density Estimate of Price')
plt.xlabel('Price')
plt.ylabel('Density')
```

```python
# Box Plot
sns.boxplot(x=airbnb['Price'], color='green')
plt.title('Box Plot of Price')
```

The data are very poorly scaled and there are SO many outliers.

```python
airbnb['price_log'] = np.log(airbnb['Price'])
```

```python
# Statistical description
description = airbnb['price_log'].describe()
print("Statistical Description:")
print(description)
```

```python
# histogram
sns.histplot(airbnb['price_log'], bins=30, kde=False, color='blue')
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
# Kernel Density Estimate (KDE)
sns.kdeplot(airbnb['price_log'], fill=True, color='orange')
plt.title('Kernel Density Estimate of Price')
plt.xlabel('Price')
plt.ylabel('Density')
```

```python
# Box Plot
sns.boxplot(x=airbnb['price_log'], color='green')
plt.title('Box Plot of Price')
```

Part 5

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=airbnb, x='Beds', y='price_log', color='blue', alpha=0.5)
plt.title('Scatter Plot of Price Log vs. Beds')
plt.xlabel('Number of Beds')
plt.ylabel('Log of Price')
plt.grid()
plt.show()
```

I see that the mean log of price is at arounf 5 to 6 for all number of beds. I see that the number of beds is commonly between 1 and 12. I also see that the variation of price seems to decrease as the number of beds increases.

```python
price_description = airbnb.groupby('Beds')['Price'].describe()
price_description
```

The mean of price seems to increase as the number of beds increases. The number of observations seems to decrease as the number of beds increases. The standard deviation seems to increase with the number of beds until 6 beds. Then, the standard deviation seems to vary a lot with the number of beds. Outliers may be playing a large role with the strange trends we see past 6 beds.
