```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

! git clone https://github.com/DS3001/visualization
```

Part 1

```python
airbnb = pd.read_csv('/content/visualization/assignment/data/airbnb_hw.csv', low_memory=False)
```

```python
airbnb['Price'] = airbnb['Price'].str.replace(',','')
airbnb['Price'] = pd.to_numeric(airbnb['Price'])
```

```python
airbnb.shape
airbnb.head(10)
```

There are 30,478 observations and 13 columns. The variables are: Host Id, Host Since, Name, Neighborhood, Property Type, Review Scores Rating, Room Type, Zipcode, Beds, Number of Records, Number of Reviews, Price, and Review Scores Rating.

Part 3

```python
crosstab_result = pd.crosstab(airbnb['Room Type'], airbnb['Property Type'])
crosstab_result
```

Part 4

```python
# Statistical description
description = airbnb['Price'].describe()
print("Statistical Description:")
print(description)
```

```python
# histogram
sns.histplot(airbnb['Price'], bins=30, kde=False, color='blue')
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
# Kernel Density Estimate (KDE)
plt.subplot(1, 3, 2)
sns.kdeplot(airbnb['Price'], fill=True, color='orange')
plt.title('Kernel Density Estimate of Price')
plt.xlabel('Price')
plt.ylabel('Density')
```

```python
# Box Plot
plt.subplot(1, 3, 3)
sns.boxplot(y=airbnb['Price'], color='green')
plt.title('Box Plot of Price')
```

```python
airbnb['price_log'] = np.log(airbnb['Price'])
```

```python
# Statistical description
description = airbnb['price_log'].describe()
print("Statistical Description:")
print(description)
```

```python
# histogram
sns.histplot(airbnb['price_log'], bins=30, kde=False, color='blue')
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
# Kernel Density Estimate (KDE)
plt.subplot(1, 3, 2)
sns.kdeplot(airbnb['price_log'], fill=True, color='orange')
plt.title('Kernel Density Estimate of Price')
plt.xlabel('Price')
plt.ylabel('Density')
```

```python
# Box Plot
plt.subplot(1, 3, 3)
sns.boxplot(y=airbnb['price_log'], color='green')
plt.title('Box Plot of Price')
```

Part 5

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=airbnb, x='Beds', y='price_log', color='blue', alpha=0.5)
plt.title('Scatter Plot of Price Log vs. Beds')
plt.xlabel('Number of Beds')
plt.ylabel('Log of Price')
plt.grid()
plt.show()
```
